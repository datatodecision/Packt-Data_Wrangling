```python
# 1. Create a list of 100 random numbers
my_list = [random.randint(0, 100) for x in range (0, 100)]
print(my_list)
```

```python
# 2. Create a new list from above list containing numbers that are divisible by 3

# Instantiate an empty list
my_list_div3 = []

# Loop through the elements of the first list and check divisibility by 3
# If true then append to new list
for i in my_list:
    if i%3 == 0:
        my_list_div3.append(i)

# Print the new list
print(my_list_div3)
```

```python
# 3. Calculate the lengths of these two lists and store the difference in a new variable
len_diff = len(my_list) - len(my_list_div3)
print(len_diff)
```

```python
# 4. Using a loop perform steps 1, 2 and 3 and find the difference variable 3 times

# Creating an empty list to capture length difference between two lists
len_diff = []

# Loop three times
# Each time create a new list with random 100 elements
# Each time create another list with elements from first list that are divisible by 3
# Each time calculate the difference in length of two above lists and append to the len_diff list
for i in range(0,3):
    my_list = [random.randint(0, 100) for x in range (0, 100)]
    my_list_div3 = []
    for i in my_list:
        if i%3 == 0:
            my_list_div3.append(i)
    
    len_diff.append(len(my_list) - len(my_list_div3))

# Print the lengths from 3 times calculations
print(len_diff)
```

```python
# 5. Calculate the arithmetic mean / common average for the lengths we have
avg_diff = sum(len_diff) / float(len(len_diff))
print(avg_diff)
```

```python
# Removing special characters and punctuations
# import re
# cleanString = re.sub('[^A-Za-z0-9]+',' ', multiline_text)
# print(cleanString)


cleaned_multiline_text = ""
for char in multiline_text:
    if char == " ":
        cleaned_multiline_text += char
    elif char.isalnum():  # using the isalnum() method of strings.
        cleaned_multiline_text += char
    else:
        cleaned_multiline_text += " "
cleaned_multiline_text
```

```python
# 4. Find all of the words in the multiline_text using split function
list_of_words = cleaned_multiline_text.split()
list_of_words
```

```python
# 5. Create a list from this list which will contain only unique words.
# Use set to get unique words
unique_words_as_list = list(set(list_of_words))
len(unique_words_as_list)
```

```python
# Use dict to do the same
unique_words_as_dict = dict.fromkeys(list_of_words)
len(list(unique_words_as_dict.keys()))
```

```python
# 6. Count the number of times the unique word has appeared in the list using the key and the value in the dict.

# looping through list of words and comparing it with it's precense in dictionary
# If present increment value for the word key by 1 else put value as 1
for word in list_of_words:
    if unique_words_as_dict[word] is None:
        unique_words_as_dict[word] = 1
    else:
        unique_words_as_dict[word] += 1
        
unique_words_as_dict
```

```python
# 7. Find the top 25 words from the unique words that you have found using slice function.

# sort the dictionary items by value in reverse order i.e. words with highest occurences will be placed om top
sorted_words_by_occurence = sorted(unique_words_as_dict.items(), key=lambda x:x[1], reverse=True)
# Pick top 25 by occurence using slice
sorted_words_by_occurence[:25]
```

### Design your own CSV parser

```python
# 1. import zip_longest from itertools
from itertools import zip_longest
```

```python
# 2. Define the return_dict_from_csv_line function so that it contains header, line, and fillvalue as None, and add it to dict
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```
